**1. What is ERPNext?**

   - **Answer:** ERPNext is an open-source, web-based Enterprise Resource Planning (ERP) software designed to help businesses manage various functions such as accounting, inventory, project management, and more.

**2. How would you customize a form in ERPNext?**

**3. Explain the difference between a Standard Report and a Custom Report in ERPNext.**

   - **Answer:** A Standard Report in ERPNext refers to pre-built reports available within the system that cover common business needs such as sales reports, purchase reports, and financial statements. On the other hand, a Custom Report is a report created by users to meet specific requirements that are not covered by the standard reports. Custom Reports can be generated using the Report Builder tool, allowing users to define filters, columns, and other parameters as needed.
